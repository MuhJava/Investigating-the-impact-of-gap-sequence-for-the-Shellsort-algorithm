```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit

df_shell_cont = pd.read_csv(r'C:\Users\mzcra\eclipse-workspace\ShellSortAlgorithim\resultsSHELL_100N_4000N_100runs5S_Continuous.csv')

# df_knuth_minus30 = pd.read_csv(r'C:\Users\mzcra\eclipse-workspace\ShellSortAlgorithim\resultsKNUTH_29524N_9840N_K9_100runs5S_-30%.csv')
column_headers = ['Seed', 'Average Comparisons', 'Average Swaps']

df_shell_cont['Average Comparisons'] = df_shell_cont['Average Comparisons'].str.replace('Comparisons: ', '')
df_shell_cont['Average Comparisons'] = df_shell_cont['Average Comparisons'].str.strip()
df_shell_cont['Average Comparisons'] = df_shell_cont['Average Comparisons'].astype(int)

N_values = range(100, 4001)

x_data = N_values  # x-coordinates (array sizes)
y_data = df_shell_cont['Average Comparisons'].values  # y-coordinates (number of comparisons)

# Triangle wave function
def triangle_wave(x, A, T, phi):
    return (4 * A / T) * np.abs(((x - T / 4 - phi) % T) - T / 2) - A

# Fit the triangle wave function
initial_guess = [516, 256, 0]  # Adjust initial guess based on your data
params, _ = curve_fit(triangle_wave, x_data, y_data, p0=initial_guess)

# Extract the fitted parameters
A, T, phi = params
print(f"Amplitude: {A}, Period: {T}, Phase Shift: {phi}")
                 
plt.figure(figsize=(10, 6))

plt.plot(N_values, df_shell_cont['Average Comparisons'], label=f'Shell', color='black')
plt.plot(x_data, triangle_wave(x_data, *params), 'r--', label='Triangle Wave')


plt.xlabel('Array Size (N)')
plt.ylabel('Comparisons')
plt.title('Comparisons vs. Array Size (N)')
plt.legend()
plt.grid(True)
plt.show()
```
